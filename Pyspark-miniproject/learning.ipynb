```
from datetime import datetime, date
from pyspark.sql import Row

df3 = spark.createDataFrame([
    Row(a=1, b=2., c='string1', d=date(2000, 1, 1), e=datetime(2000, 1, 1, 12, 0)),
    Row(a=2, b=3., c='string2', d=date(2000, 2, 1), e=datetime(2000, 1, 2, 12, 0)),
    Row(a=4, b=5., c='string3', d=date(2000, 3, 1), e=datetime(2000, 1, 3, 12, 0))
])
display(df3)
df3.show()
df3.printSchema()
```

```
df = spark.createDataFrame([
    (1, 2., 'string1', date(2000, 1, 1), datetime(2000, 1, 1, 12, 0)),
    (2, 3., 'string2', date(2000, 2, 1), datetime(2000, 1, 2, 12, 0)),
    (3, 4., 'string3', date(2000, 3, 1), datetime(2000, 1, 3, 12, 0))
], schema='a long, b double, c string, d date, e timestamp')
display(df)
```

```
is_first_row_header = True
matches_df = spark.read.format("csv").option("header", is_first_row_header).load("/FileStore/tables/matches.csv")
# display(matches_df)

inferSchema = True
matches_df2 = spark.read.format("csv").option("inferSchema", inferSchema).option("header", is_first_row_header).load("/FileStore/tables/matches.csv")
matches_df2.printSchema()
```

```
matches_df2.filter(matches_df2.id=="1").show()
matches_df2.where("id=1 or id=2").show()
# Both works the same
```

```
from pyspark.sql.functions import col
matches_df2.sort(matches_df2.win_by_wickets.desc()).show()
matches_df2.orderBy(col("win_by_wickets").desc()).show()
# Both are the same
```
